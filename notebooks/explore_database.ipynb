```python
print("Check that tables exist")
df_tables = db.query(
    """
    SELECT table_name
    FROM information_schema.tables
    WHERE table_schema = 'public'
      AND table_type = 'BASE TABLE'
    ORDER BY table_name;
"""
)
df_tables
```

```python
print("Total number of reviews")
df_total_reviews = db.query(
    """
    SELECT COUNT(*) AS total_reviews
    FROM reviews;
"""
)
df_total_reviews
```

```python
print("Reviews per bank")
df_reviews_per_bank = db.query(
    """
    SELECT b.bank_name, COUNT(r.review_id) AS review_count
    FROM reviews r
    JOIN banks b ON r.bank_id = b.bank_id
    GROUP BY b.bank_name
    ORDER BY review_count DESC;
"""
)
df_reviews_per_bank
```

```python
print("Average rating per bank")
df_avg_rating = db.query(
    """
    SELECT b.bank_name, AVG(r.rating) AS avg_rating
    FROM reviews r
    JOIN banks b ON r.bank_id = b.bank_id
    GROUP BY b.bank_name
    ORDER BY avg_rating DESC;
"""
)
df_avg_rating
```
